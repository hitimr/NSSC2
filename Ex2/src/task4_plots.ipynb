```python
#!/usr/bin/env python3
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy
import pandas as pd

#include parent folder
import os, sys, inspect

currentdir = os.path.dirname(
    os.path.abspath(inspect.getfile(inspect.currentframe())))
parentdir = os.path.dirname(currentdir)
sys.path.insert(0, parentdir)
from src.domain import * 
```

# sigma=1, dt=0.01; it=100

```python
#set up domain
rho = 0.8
N = 200
L = (float(N/rho)**(1/3))
dt = 0.01
sigma = 1
spread = 1/(3*N**(1/3))
domain = Domain()
domain.fill(N, L, spread, sigma)
domain.minimizeEnergy()

#compute energy
Ekin_vals0 = []
Epot_vals0 = []
it =[]
for i in range(100):
    it.append(i)
    domain.verlet_advance(dt)
    Ekin_vals0.append(domain.Ekin())
    Epot_vals0.append(domain.Epot())
Ekin_vals0 = np.array(Ekin_vals0)
Epot_vals0 = np.array(Epot_vals0)

#plot
fig = plt.figure(figsize = (7,5))
plt.plot(it,Ekin_vals0,label="E_kin")
plt.plot(it,Epot_vals0,label="E_pot")
plt.plot(it,Ekin_vals0+Epot_vals0,label="E_tot")
plt.legend()
plt.xlabel("timesteps")
plt.title("Energy, sigma=1, N=200, dt=0.01")
plt.grid()
#plt.show()
plt.savefig('plots/energy_s10_n200_dr001.pdf')
```
