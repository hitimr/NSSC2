```python


nx = 10
ny = 10
L = 10

# values in polar coordinateL
a, b = 1.25, 1
r_vals = np.linspace(2*L, L, nx)
phi_vals = np.linspace(a*np.pi, b*np.pi, ny) 

# generate polar mesh
r, phi = np.meshgrid(r_vals, phi_vals)


x,y = [], []
for i in range(len(r)):
    x.append(r[i]*np.cos(phi[i]))
    y.append(r[i]*np.sin(phi[i]))

x = np.array(x).ravel() + 2*L
y = np.array(y).ravel()
plt.xlim([-1, 15])
plt.ylim([-15,1])
plt.scatter(x,y)
```

```python
plt.scatter(x[0:10],y[0:10], label="N0-N9")
plt.scatter(x[10:20], y[10:20], label="N10-N19")
plt.scatter(x[20:25], y[20:25], label="N20-N24")
plt.xlim([-1, 15])
plt.ylim([-15,1])
plt.legend()
```
