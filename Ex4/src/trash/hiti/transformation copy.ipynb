```python


nx = 3
ny = 3
L = 10

# values in polar coordinateL
a, b = 1.25, 1
r_vals = np.linspace(1, 0, nx)
phi_vals = np.linspace(a*np.pi, b*np.pi, ny) 

# generate polar mesh
r, phi = np.meshgrid(r_vals, phi_vals)
print(phi)
print(r)

x,y = [], []
for i in range(len(r)):
    x.append(r[i]*np.cos(phi[i]))
    y.append(r[i]*np.sin(phi[i]))


x = np.array(x).ravel() + 2*L
y = np.array(y).ravel()
plt.xlim([-1, 15])
plt.ylim([-15,1])
plt.scatter(x,y)

# print(x)
```
