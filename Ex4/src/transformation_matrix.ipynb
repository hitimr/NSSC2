```python
import numpy as np
import matplotlib.pyplot as plt

n = 10
y, x = np.meshgrid(np.linspace(0,1,n),np.linspace(0,1,n))

z = np.sin(x)**10 + np.cos(10 + y*x) * np.cos(x)

plt.pcolor(x,y,z)
```

```python
nx, ny = (5,5)
x = np.linspace(0, 1, nx)
y = np.linspace(0, 1, ny)
xv, yv = np.meshgrid(x,y)

zv = xv*yv

plt.pcolor(xv,yv, zv)
```
