```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt("leveling_analysis2b.txt", dtype=None, delimiter=",")
print(data)
data6 = np.delete(data, 1, axis=0)
```

```python
cities6 = []
population79=[]
population19=[]
absolutegrowth=[]


for i in range(len(data6)):
    cities6.append(data6[i][0])
    population79.append(data6[i][1])
    population19.append(data6[i][2])
    absolutegrowth.append(data6[i][3])
    
print(absolutegrowth)
```

```python

plt.bar(cities6, population79, color='green')
plt.xlabel("city")
plt.ylabel("population")

plt.show()
#plt.savefig("asdf.pdf")
```

```python
fig, axs = plt.subplots(1, 2,sharey=True, tight_layout=True)

#set number of bins
axs[0].hist(population79, bins=6)
axs[1].hist(absolutegrowth, bins=6)
```
